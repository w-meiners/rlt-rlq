# Aufgabe 3

a) Erforderlicher Außenluftvolumenstrom

$$
  \dot V_{au} = \dfrac{\dot V_{sch}}{k_{zul}-k_{au}}
$$

```python
k_zul = 1000e-6 # 1000 ppM
k_au = 400e-6 # 400 ppM

dV_co2 = 120e-3 # m**3/h

dV_au = dV_co2/(k_zul-k_au)
dV_au # m**3/h
```

Es werden 200 Kubikmeter pro Stunde Außenluft benötigt.

b) Luftwechselrate des Raumes

$$
  \beta = \dfrac{\dot V_{au}}{V_{ra}}
$$

```python
V_ra = 40*2.5 # m**3

beta = dV_au/V_ra
beta # 1/h
```

Es ergibt sich ein Luftwechsel von $2\,h^{-1}$.

c) Darstellung der Schadstoffkonzentration $k(t)$ bei der Entlüftung.

$$
  k(t) = 
  k_\infty + (k_0 - k_\infty)\cdot\mathrm{e}^{-\beta\,(t-t_0)}
$$

Der Raum ist zunächst von Personen belegt (d.h. $k = 1000\,\mathrm{ppM}$)

Dieser Raum wird nun entlüftet. Dabei sinkt die CO_2-Konzentration schließlich auf $k = 400\,\mathrm{ppM}$.

Deshalb ist 

$$
  k_0 = 1000\,\mathrm{ppm}
$$

und 

$$
  k_\infty = 400\,\mathrm{ppM}
$$

$$
  k(t) = 
  k_\infty + (k_0 - k_\infty)\cdot\mathrm{e}^{-\beta\,(t-t_0)}
$$

```python
import pandas as pd
import numpy as np

%config InlineBackend.figure_format='retina' # nur MacBook Pro
```

```python
lt = np.linspace(0,4) # Das Zeitintervall

k_inf = 400e-6 # 400 ppM
k_0 = 1000e-6 # 1000 ppM

df = pd.DataFrame(
    {
        't': lt,
        'k': (k_inf + (k_0-k_inf)*np.exp(-beta*lt))*1e6 # in ppM
    }
)

df.head().T
```

```python
ax = df.plot(x='t',y='k',label="$k=k(t)$")

ax.set(
    ylim=(0,1200),ylabel='CO_2-Konzentration $k$ in ppM',
    xlim=(0,4), xlabel='Zeit $t$ in Stundent'
)

ax.axhline(k_au*1e6,c='r')

ax.grid()
```
