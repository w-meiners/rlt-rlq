```python
from matplotlib import pyplot as plt, ticker as tk
%matplotlib inline
%config InlineBackend.figure_format='retina'

import numpy as np
import pandas as pd

from sympy import *
init_printing()

from fun_expr import Function_from_Expression as FE
```

```python
k_0 = 400
k_inf = 1100
beta = 1/20

t = Symbol('t')

k = FE(t, k_inf + (k_0-k_inf)*exp(-beta*t),name='k')
k.equation()
```

```python
fig,ax = plt.subplots()

ax.axhline(k_inf,c='r')
lx = np.linspace(0,120)
ax.plot(lx,k.lambdified(lx))

ax.set(
    xlim=(0,120), xlabel='Zeit $t$ in Minuten',
    ylim=(0,1200), ylabel='CO_2-Konzentration in ppM'
)

ax.grid(lw=0.25,c='k')
plt.savefig('U03-A02-grafik.pdf',bbox_inches='tight')
```
