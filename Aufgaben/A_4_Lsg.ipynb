Da die Bedingungen die gleichen sind, wie in Aufgabe 3, muss auch der Außenluftvolumenstrom der gleiche sein, wie dort. 

b) Verlauf der Schadstoffkonzentration im Raum

```python
V_ra = 40*2.5 # m**3

k_au = k_0 = 400e-6 # 400 ppM
k_zul = 1000e-6 # 1000 ppM

dV_co2 = 4*30e-3 # m**3/h
dV_au = dV_co2/(k_zul-k_au)

beta = dV_au/V_ra
beta
```

```python
%config InlineBackend.figure_format='retina' # nur MacBook Pro

import numpy as np
import pandas as pd
```

```python
lt = np.linspace(0,4) # Zeitintervall von vier Stunden

df = pd.DataFrame(
    {
        't': lt,
        'k': (k_zul + (k_0-k_zul)*np.exp(-beta*lt))*1e6 # in ppM
    }
)
df.head().T
```

```python
ax = df.plot(x='t',y='k',label='$k=k(t)$')
ax.axhline(k_zul*1e6,c='r') # in ppM

ax.set(
    xlim=(0,4),xlabel='Zeit $t$ in Stunden',
    ylim=(0,1200),ylabel='CO_2-Konzentration $k$ in ppM',
)
ax.grid()
```
